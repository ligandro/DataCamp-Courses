```python
#From a  dictionarie OF lists
data = {'Name':[5, 10, 15, 50],
        'Age':[20, 21, 19, 18]}  #each key represents a column

# Create the pandas DataFrame
df = pd.DataFrame(data, columns = ['Name', 'Age'])
df
```

#### Adding error-bars to a bar chart

```python
fig, ax = plt.subplots()

# Add a bar for the rowing "Height" column mean/std
ax.bar("Rowing", df["Name"].mean(), yerr=df["Name"].std())

# Add a bar for the gymnastics "Height" column mean/std
ax.bar("Gymnastics", df["Age"].mean(), yerr=df["Age"].std())

# Label the y-axis
ax.set_ylabel("Height (cm)")

plt.show()
```

#### Adding error-bars to a plot

```python
fig, ax = plt.subplots()


# Implementation of matplotlib function
import numpy as np
# example data
xval = np.arange(0.1, 4, 0.5)
yval = np.exp(-xval)
  
ax.errorbar(xval, yval, xerr = 0.4, yerr = 0.5)
  
ax.set_title('matplotlib.pyplot.errorbar() function Example')

plt.show()
```

### Creating boxplots

#### helps finding out outliers

```python
fig, ax = plt.subplots()

# Add a boxplot for the "Height" column in the DataFrames
ax.boxplot([df["Name"], df["Age"]])

# Add x-axis tick labels:
ax.set_xticklabels(["Rowing", "Gymnastics"])

# Add a y-axis label
ax.set_ylabel("Height (cm)")

plt.show()
```

## ->Automating figures from data

#### This means that you can write functions and programs that automatically adjust what they are doing based on the input data. Why would you want to automate figure creation based on the data? Automation makes it easier to do more. It also allows you to be faster. This is one of the major benefits of using a programming language like Python and software libraries such as Matplotlib,

```python
#From a  dictionarie OF lists
data = {'Name':["A","B", "C", "D"],
        'Age':[20, 21, 19, 18]}  #each key represents a column

# Create the pandas DataFrame
df = pd.DataFrame(data, columns = ['Name', 'Age'])
df
```

```python
# Extract the "Sport" column
sports_column = df["Name"]

# Find the unique values of the "Sport" column
sports = sports_column.unique()

# Print out the unique sports values
print(sports)
```

```python
fig, ax = plt.subplots()

# Loop over the different sports branches
for sport in sports:
  # Extract the rows only for this sport
  sport_df = df[df["Name"] == sport]
  # Add a bar for the "Weight" mean with std y error bar
  ax.bar(sport, sport_df["Age"].mean(), yerr=sport_df["Age"].std())

ax.set_ylabel("Weight")
ax.set_xticklabels(sports, rotation=90)
```
